### 01) WAP to demonstrate the use of Pie chart.

```python
x=[10,20,30,40]
y=["food","travel","accomodation","shopping"]
c=["r","g","y","m"]
ex=[0.5,0.0,0.0,0.0]
plt.pie(x,labels=y,colors=c,shadow=True,radius=1.3,autopct="%0.2f%%",labeldistance=1.2,
        textprops={"fontsize":15,"color":"pink"},explode=ex)
plt.title("Expensis")
colors=c
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
x=["python","c","c++","java"]
y=[85,70,60,80]
c=["r","g","y","m"]
plt.bar(x,y,color=c,align='edge',edgecolor='b',alpha=0.5,)
plt.title("Popularity of langauges")
plt.xlabel("langauges")
plt.ylabel("popularity")
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x=["python","c","c++","java"]
y=[85,70,60,80]
c=["r","g","y","m"]
plt.bar(x,y,color=c,align='edge',edgecolor='b',alpha=0.5,)
plt.title("Popularity of langauges")
plt.xlabel("langauges")
plt.ylabel("popularity")
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
addlabels(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
c=["r","g","y","m","b"]
s=[150,70,190,50,120]
plt.scatter(x,y,color=c,marker='*',s=s,edgecolors=["g","r","b","y","m"])
plt.title("CPI && SPI")
plt.xlabel("CPI")
plt.ylabel("SPI")
plt.show()
```
